## Machine Learning Algorithms applied to Divorce Data

```python
# Import dependencies

import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os

```

```python
# read data from csv file into pandas data frame
df = pd.read_csv("divorce_data.csv")
df.head()
```

### Linear Support Vector Classification

```python
target = df["Class"]
target_names = ["Divorced", "Married"]
```

```python
data = df.drop("Class", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(data, target)
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
print('Test Acc: %.5f' % model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
